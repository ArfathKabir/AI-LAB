```python
def clean(floor, row, coln):
    i, j, m, n = row, coln, len(floor), len(floor[0])
    right = down = True
    cleaned = [not any(f) for f in floor]
    if cleaned.count(True) >= 3:
           print_floor(floor, i, j)
           return
    while not all(cleaned):
        while any(floor[i]):
            print_floor(floor, i, j)
            if floor[i][j]:
                floor[i][j] = 0
                print_floor(floor, i, j)
            if not any(floor[i]):
                cleaned[i] = True
                break
            if j == n - 1:
                j -= 1
                right = False
            elif j == 0:
                j += 1
                right = True
            else:
                j += 1 if right else -1
        if cleaned.count(True) >= 3 :
           break
        if i == m - 1:
            i -= 1
            down = False
        elif i == 0:
            i += 1
            down = True
        else:
            i += 1 if down else -1
        if cleaned[i]:
            print_floor(floor, i, j)
def print_floor(floor, row, col):
    for r in range(len(floor)):
        for c in range(len(floor[r])):
            if r == row and c == col:
                print(f" >{floor[r][c]}< ", end = ' ')
            else:
                print(f"  {floor[r][c]}  ", end = ' ')
        print(end = '\n')
    print(end = '\n' )            
```

```python
floor = [[1, 0, 0, 0],
         [0, 1, 0, 1],
         [1, 0, 1, 1],
         [1, 0, 0, 0]]
clean(floor, 0, 0)
```
